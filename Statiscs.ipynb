```python
#Q10:  Explain nominal, ordinal, interval, and ratio levels of measurement
#Ans: Nominal Scale: This is the most basic level. Data is categorized into distinct, mutually exclusive groups.   
#                    There's no inherent order or ranking among the categories.
#Ordinal Scale: Data is categorized and ranked in a specific order.   
#               The intervals between the ranks are not necessarily equal.   
#Interval Scale: Data is categorized, ranked, and has equal intervals between values.   
#                There's no true zero point, meaning zero doesn't indicate the absence of the measured attribute.
#Ratio Scale: This is the highest level of measurement.   
#             Data has all the properties of interval data, plus a true zero point.
#             Ratios between values are meaningful.
```

```python
#Q11. What is the measure of central tendency
#Ans: Measures of central tendency represent the typical value in a dataset, summarizing the data's center. Common measures include the mean, median, and mode.
```

```python
#Q12. Define mean, median, and mode
#Ans: Mean: The average of all values in a dataset, calculated by summing all values and dividing by the number of values.
#Median: The middle value in a dataset when the values are arranged in ascending or descending order.If there's an even number of values, it's the average of the two middle values
#Mode: The value that appears most frequently in a dataset.
```

```python
#Q13. What is the significance of the measure of central tendency
#Ans:The significance of measures of central tendency lies in their ability to:
#Summarize data: They provide a single, representative value that captures the essence of an entire dataset, making it easier to understand and compare.   
#Identify typical values: They help determine what is "typical" or "average" within a distribution, revealing patterns and trends.
```

```python
#Q14.What is variance, and how is it calculated
#Ans:Variance measures how spread out the data points are from the mean. It's calculated by averaging the squared differences between each data point and the mean.
```

```python
#Q15. What is standard deviation, and why is it important
#Ans: Standard deviation measures the amount of variation or dispersion of a set of values. It's important because it quantifies how much data points deviate from the mean, providing a clearer picture of data spread and reliability.
```

```python
#Q16.Define and explain the term range in statistics
#Ans: The range is a simple measure of dispersion. It represents the difference between the largest and smallest values in a dataset
#Range = Maximum Value - Minimum Value
```

```python
#Q17.What is the difference between variance and standard deviation
#Ans:Variance is the average of the squared differences from the mean, while standard deviation is the square root of the variance. Standard deviation is in the same units as the data, making it more interpretable.
```

```python
#Q18.  What is skewness in a dataset
#Ans:Skewness in a dataset refers to the asymmetry or lack of symmetry in its distribution. It indicates whether the data is concentrated more on one side of the mean than the other.
```

```python
#Q19.What does it mean if a dataset is positively or negatively skewed
#Ans:Positive skewness means the tail extends right, with most data on the left; negative skewness means the tail extends left, with most data on the right.
```

```python
#Q20. Define and explain kurtosis
#Ans: Kurtosis measures the degree to which a distribution's tails are heavy or light relative to a normal distribution. It essentially quantifies the shape of the distribution's tails.   
```

```python
#Q21.What is the purpose of covariance
#Ans:Covariance measures the degree to which two variables change together. Its purpose is to indicate the direction (positive or negative) of the linear relationship between them.
```

```python
#Q22.  What does correlation measure in statistic
#Ans:Correlation measures the strength and direction of a linear relationship between two variables. It indicates how closely they move together, ranging from -1 (perfect negative correlation) to +1 (perfect positive correlation).
```

```python
#Q23.What is the difference between covariance and correlation
#Ans:Covariance indicates the direction of a linear relationship, but its scale is unstandardized. Correlation standardizes covariance, ranging from -1 to +1, showing both direction and strength. Correlation is more interpretable for comparing relationships across datasets.
```

```python
#Q8. Generate a dataset and demonstrate positive and negative skewness
from scipy.stats import skew
positive_skew_data = np.random.exponential(scale=2, size=100)
negative_skew_data = np.random.beta(a=5, b=2, size=100) * 10

pos_skewness = skew(positive_skew_data)
neg_skewness = skew(negative_skew_data)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(positive_skew_data, bins=30, kde=True, color='blue', ax=axes[0])
axes[0].set_title(f"Positive Skewness (Right Skew)\nSkewness: {pos_skewness:.2f}")

sns.histplot(negative_skew_data, bins=30, kde=True, color='red', ax=axes[1])
axes[1].set_title(f"Negative Skewness (Left Skew)\nSkewness: {neg_skewness:.2f}")

plt.show()
```

```python
#Q13. Calculate the mean, median, and mode of grouped data
import numpy as np
from scipy import stats

midpoints = [5, 15, 25, 35, 45]
frequencies = [5, 8, 15, 10, 7]

mean_value = np.average(midpoints, weights=frequencies)

cumulative_frequencies = np.cumsum(frequencies)
N = sum(frequencies)
median_class_index = np.where(cumulative_frequencies >= N/2)[0][0]  
median_value = midpoints[median_class_index]

mode_index = np.argmax(frequencies)
mode_value = midpoints[mode_index] 

print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value:.2f}")
print(f"Mode: {mode_value:.2f}")
```

```python
#Q23.Visualize skewness and kurtosis using Python libraries like matplotlib or seaborn
from scipy.stats import skew, kurtosis
normal_data = np.random.rand(50)
positively_skewed = np.random.exponential(scale=10, size=1000)  
negatively_skewed = -np.random.exponential(scale=10, size=1000) + 100  

def plot_distribution(data, title):
    plt.figure(figsize=(6, 4))
    sns.histplot(data, bins=30, kde=True, color='skyblue')
    plt.title(f"{title}\nSkewness: {skew(data):.2f} | Kurtosis: {kurtosis(data):.2f}")
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.show()

plot_distribution(normal_data, "Normal Distribution")
plot_distribution(positively_skewed, "Positively Skewed Distribution")
plot_distribution(negatively_skewed, "Negatively Skewed Distribution")
```
